### Código de Criação e Edição do Banco de Dados

```python
# Bibliotecas

import sqlite3
import pandas as pd
```

```python
# Carregar o dataset
df = pd.read_csv('GlobalLandTemperaturesByCountry.csv')

# Exibir as primeiras linhas do dataset
print("Antes do tratamento de dados:")
print(df.head())
```

```python
# Tratamento de dados: Remover valores null e NaN
df.dropna(inplace=True)

# Exibir as primeiras linhas do dataset após o tratamento de dados
print("Depois do tratamento de dados:")
print(df.head())
```

```python
# Conectar ao banco de dados SQLite
conn = sqlite3.connect('global_temperatures.db')
c = conn.cursor()

# Criar a tabela
c.execute('''
    CREATE TABLE IF NOT EXISTS temperatures (
        dt TEXT,
        AverageTemperature REAL,
        AverageTemperatureUncertainty REAL,
        Country TEXT
    )
''')

# Inserir os dados no banco de dados
df.to_sql('temperatures', conn, if_exists='replace', index=False)

# Confirmar a criação da tabela
c.execute('SELECT * FROM temperatures LIMIT 5')
print(c.fetchall())

# Fechar a conexão com o banco de dados
conn.close()
```
